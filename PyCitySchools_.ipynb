# PyCity Schools Analysis

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending per student actually (\$645-675) underperformed compared to schools with smaller budgets (<\$585 per student).

* As a whole, smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%).

* As a whole, charter schools out-performed the public district schools across all metrics. However, more analysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school.

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

#Student ID,student_name,gender,grade,school_name,reading_score,math_score
#School ID,school_name,type,size,budget
```

* Calculate the total number of schools

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

```python
# Total Schools
total_schools = school_data["school_name"].count()
#print(total_schools)


# Total Students
total_students = student_data["school_name"].count()
print(total_students)

# Total Budget
total_budget = school_data["budget"].sum()
print(total_budget)

# Average Math Score
avg_math_score = student_data["math_score"].mean()
print(avg_math_score)

# Average Reading Score
avg_reading_score = student_data["reading_score"].mean()
print(avg_reading_score)
```

```python
#Overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
overall_pass = (avg_math_score + avg_reading_score)/2
print(overall_pass)
```

```python
 #Percentage of students with a passing math score (70 or greater)
## % Passing Math based on 70
math_pass = student_data.loc[(student_data["math_score"] >= 70)]

count_pass_math = math_pass["math_score"].count()
#print(count_pass_math)

per_math_pass = (count_pass_math/total_students)*100
print(per_math_pass)
```

```python
 #Percentage of students with a passing reading score (70 or greater)
read_pass = student_data.loc[(student_data["reading_score"] >= 70)]

count_pass_read = read_pass["reading_score"].count()
#print(count_pass_read)

per_read_pass = (count_pass_read/total_students)*100
print(per_read_pass)
```

```python
# Create a dataframe to hold the above results
# Optional: give the displayed data cleaner formatting
district_summary = {"Total Schools" : total_schools,
                   "Total Students" : total_students,
                   "Total Budget" : total_budget,
                   "Average Math Score" : avg_math_score,
                   "Average Reading Score" : avg_reading_score,
                    "% Passing Math" : per_math_pass,
                   "% Passing Reading" : per_read_pass,
                   "% Overall Passing" : overall_pass 
                   }


district_summary_df = pd.DataFrame([district_summary])   

district_summary_df =  district_summary_df[["Total Schools",
                   "Total Students","Total Budget","Average Math Score", "Average Reading Score",
                   "% Passing Math", "% Passing Reading", "% Overall Passing" ]]
district_summary_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
scores_by_school_spending = school_data_complete.copy()
scores_by_school_spending.columns
```

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
scores_by_school_spending['spending_bins'] = pd.cut(scores_by_school_spending['budget']/scores_by_school_spending['size'], spending_bins, labels=group_names)

#group by spending
by_spending = scores_by_school_spending.groupby('spending_bins')

#calculations
avg_math = by_spending['math_score'].mean()
avg_read = by_spending['reading_score'].mean()
pass_math = scores_by_school_spending[scores_by_school_spending['math_score'] >= 70].groupby('spending_bins')['Student ID'].count()/by_spending['Student ID'].count()*100
pass_read = scores_by_school_spending[scores_by_school_spending['reading_score'] >= 70].groupby('spending_bins')['Student ID'].count()/by_spending['Student ID'].count()*100
overall = (pass_math + pass_read )/2

scores_by_school_spending = pd.DataFrame({ "Average Math Score": avg_math, "Average Reading Score": avg_read,
                                           '% Passing Math': pass_math, '% Passing Reading': pass_read,
                                           "Overall Passing Rate": overall
                                          })
            


##reorder columns
scores_by_school_spending = scores_by_school_spending[[
    "Average Math Score",
    "Average Reading Score",
    '% Passing Math',
    '% Passing Reading',
    "Overall Passing Rate"
]]
 
    

scores_by_school_spending.index.name = "Spending Ranges"
scores_by_school_spending

```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
scores_by_school_size = school_data_complete.copy()
scores_by_school_size.columns
```

```python
# Sample bins. Feel free to create your own bins.
bins = [0, 1000, 2000, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
scores_by_school_size['size_bins'] = pd.cut(school_data_complete['size'], bins, labels = group_names)

by_size = scores_by_school_size.groupby('size_bins')

avg_math = by_size['math_score'].mean()
avg_read = by_size['reading_score'].mean()
pass_math = scores_by_school_size[scores_by_school_size['math_score'] >= 70].groupby('size_bins')['Student ID'].count()/by_size['Student ID'].count()*100
pass_read = scores_by_school_size[scores_by_school_size['reading_score'] >= 70].groupby('size_bins')['Student ID'].count()/by_size['Student ID'].count()*100
overall = (pass_math + pass_read )/2

scores_by_school_size = pd.DataFrame({"Average Math Score": avg_math,
                                      "Average Reading Score": avg_read,
                                      '% Passing Math': pass_math,
                                      '% Passing Reading': pass_read,
                                      "Overall Passing Rate": overall
                                     })

scores_by_school_size = scores_by_school_size[["Average Math Score","Average Reading Score",'% Passing Math',
                                 '% Passing Reading',"Overall Passing Rate"]]

scores_by_school_size.index.name = "School Size"
scores_by_school_size
```
